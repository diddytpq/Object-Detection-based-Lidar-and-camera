```python
import time
import numpy as np

def count_point_in_box(box_corner, pc):
    t0 = time.time()

    point_count = 0

    x_min = box_corner[:,0].min()
    y_min = box_corner[:,1].min()
    z_min = box_corner[:,2].min()
    x_max = box_corner[:,0].max()
    y_max = box_corner[:,1].max()
    z_max = box_corner[:,2].max()

    # for (x,y,z,r) in pc:
    #     if x_min < x < x_max:
    #         if y_min < y < y_max:
    #             if z_min < z < z_max:
    #                 point_count += 1


    x_count = np.where((x_min < pc[:,0]) & (pc[:,0] < x_max) & (y_min < pc[:,1]) & (pc[:,1] < y_max) & (z_min < pc[:,2]) & (pc[:,2] < z_max) )
    

    print(len(x_count[0]))
    print(point_count)



    print(time.time() - t0)


    return point_count
```
